# Lab 7- Deep Learning Model

This lab is meant to get you started in using Keras to design Deep Neural Networks. The goal here is to simply repeat your previous lab, but with DNNs.

Let's start with reading the data, like before:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

filename="../Lab5/SUSY.csv"
VarNames=["signal", "l_1_pT", "l_1_eta","l_1_phi", "l_2_pT", "l_2_eta", "l_2_phi", "MET", "MET_phi", "MET_rel", "axial_MET", "M_R", "M_TR_2", "R", "MT2", "S_R", "M_Delta_R", "dPhi_r_b", "cos_theta_r1"]
RawNames=["l_1_pT", "l_1_eta","l_1_phi", "l_2_pT", "l_2_eta", "l_2_phi","MET", "MET_phi", "MET_rel", "axial_MET"]
FeatureNames=["M_R", "M_TR_2", "R", "MT2", "S_R", "M_Delta_R", "dPhi_r_b", "cos_theta_r1"]

df = pd.read_csv(filename, dtype='float64', names=VarNames)
```

Now lets define training and test samples. Note that DNNs take very long to train, so for testing purposes we will use only about 10% of the 5 million events in the training/validation sample. Once you get everything working, make the final version of your plots with the full sample. 

Also note that Keras had trouble with the Pandas tensors, so after doing all of the nice manipulation that Pandas enables, we convert the Tensor to a regular numpy tensor.

```python
N_Max=550000
N_Train=500000

Train_Sample=df[:N_Train]
Test_Sample=df[N_Train:N_Max]

X_Train=np.array(Train_Sample[VarNames[1:]])
y_Train=np.array(Train_Sample["signal"])

X_Test=np.array(Test_Sample[VarNames[1:]])
y_Test=np.array(Test_Sample["signal"])
```

## Exercise 7.1

You will need to create several models and make sure they are properly trained. Write a function that takes this history and plots the values versus epoch. For every model that you train in the remainder of this lab, assess:

* Has you model's performance plateaued? If not train for more epochs. 
* Compare the performance on training versus test sample. Are you over training?

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout
import matplotlib.pyplot as plt
import numpy as np
```

**model 1**

```python
model = Sequential()
model.add(Dense(12, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(X_Train,y_Train, validation_split=0.1, epochs=8)
```

```python
model.evaluate(X_Test,y_Test)
score1 = model.evaluate(X_Test, y_Test, verbose=1)
print('Model Test accuracy:', score1[1])
print('Model Test loss:', score1[0])
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='lower right')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='lower right')
plt.show()
```

**model 2**

```python
plt.plot(history2.history['accuracy'])
plt.plot(history2.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='lower right')
plt.show()
```

**model 3**

**model 4**

```python
model4 = Sequential()
model4.add(Dense(17, kernel_initializer="random_uniform", activation='tanh'))
model4.add(Dense(36, kernel_initializer="random_uniform", activation='relu'))
model4.add(Dropout(0.5))
model4.add(Dense(72, kernel_initializer="random_uniform", activation='relu'))
model4.add(Dropout(0.5))
model4.add(Dense(36, kernel_initializer="random_uniform", activation='relu'))
model4.add(Dropout(0.5))
model4.add(Dense(1,kernel_initializer="random_uniform", activation='sigmoid'))
```

```python
model4.compile(loss = 'binary_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

```python
history4 = model4.fit(X_Train,y_Train, validation_data=(X_Test,y_Test), epochs=10, batch_size=128, shuffle=True)
```

```python
model4.evaluate(X_Test,y_Test)
score4 = model4.evaluate(X_Test, y_Test, verbose=1)
print('Model Test accuracy:', score4[1])
print('Model Test loss:', score4[0])
```

```python
plt.plot(history4.history['accuracy'])
plt.plot(history4.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='lower right')
plt.show()
```

```python
plt.plot(history4.history['loss'])
plt.plot(history4.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='lower right')
plt.show()
```

## Exercise 7.2

Following the original paper (see lab 5), make a comparison of the performance (using ROC curves and AUC) between models trained with raw, features, and raw+features data.

```python
X_Test_raw=np.array(Test_Sample[RawNames])
y_Test_raw=np.array(Test_Sample["signal"])
```

```python
X_Test_fea=np.array(Test_Sample[FeatureNames])
y_Test_fea=np.array(Test_Sample["signal"])
```

```python
X_Test=np.array(Test_Sample[VarNames[1:]])
y_Test=np.array(Test_Sample["signal"])
```

```python
from sklearn.metrics import roc_curve,auc

predict_prob_raw=model.predict(X_Test_raw)

predict_classes_raw=np.argmax(predict_prob_raw,axis=1)

fpr_keras_raw, tpr_keras_raw, thresholds_keras_raw = roc_curve(y_Test_raw, predict_classes_raw)
```

```python
auc_keras_raw = auc(fpr_keras_raw, tpr_keras_raw)
```

```python
plt.figure()
plt.plot([0, 1], [0, 1], 'k--')
plt.title('Receiver Operating Characteristic (Raw)')
plt.plot(fpr_keras_raw, tpr_keras_raw, label='(area = {:.3f})'.format(auc_keras_raw))
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.legend(loc='best')
plt.show()
```

```python
predict_prob_fea=model.predict(X_Test_fea)

predict_classes_fea=np.argmax(predict_prob_fea,axis=1)

fpr_keras_fea, tpr_keras_fea, thresholds_keras_fea = roc_curve(y_Test_fea, predict_classes_fea)
```

```python
auc_keras_fea = auc(fpr_keras_fea, tpr_keras_fea)
```

```python
plt.figure()
plt.plot([0, 1], [0, 1], 'k--')
plt.title('Receiver Operating Characteristic (Features)')
plt.plot(fpr_keras_fea, tpr_keras_fea, label='(area = {:.3f})'.format(auc_keras_fea))
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.legend(loc='best')
plt.show()
```

```python
predict_prob=model.predict(X_Test)

predict_classes=np.argmax(predict_prob,axis=1)

fpr_keras, tpr_keras, thresholds_keras = roc_curve(y_Test, predict_classes)
```

```python
auc_keras = auc(fpr_keras, tpr_keras)
```

```python
plt.figure()
plt.plot([0, 1], [0, 1], 'k--')
plt.title('Receiver Operating Characteristic (Overall)')
plt.plot(fpr_keras, tpr_keras, label='(area = {:.3f})'.format(auc_keras))
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.legend(loc='best')
plt.show()
```
