### Exercise 3: Training a Classifier

Lets' repeat what we did manually in the previous lab using scikit-learn. We'll use a LDA classifier, which we can instanciate as follows:

```python
import sklearn.discriminant_analysis as DA
Fisher=DA.LinearDiscriminantAnalysis()
```

As discussed in the lecture, to properly formulate our problem, we'll have to:

* Define the inputs (X) vs outputs (Y)
* Designate training vs testing samples (in order to get a unbias assessment of the performance of Machine Learning algorithms)

for example, here we'll take use 4M events for training and the remainder for testing.

```python
N_Train=4000000

Train_Sample=df[:N_Train]
Test_Sample=df[N_Train:]

X_Train=Train_Sample[VarNames[1:]]
y_Train=Train_Sample["signal"]

X_Test=Test_Sample[VarNames[1:]]
y_Test=Test_Sample["signal"]

Test_sig=Test_Sample[Test_Sample.signal==1]
Test_bkg=Test_Sample[Test_Sample.signal==0]
```

We can train the classifier as follow:

```python
Fisher.fit(X_Train,y_Train)
```

We can plot the output, comparing signal and background:

```python
plt.figure()
plt.hist(Fisher.decision_function(Test_sig[VarNames[1:]]),bins=100,histtype="step", color="blue", label="signal",stacked=True)
plt.hist(Fisher.decision_function(Test_bkg[VarNames[1:]]),bins=100,histtype="step", color="red", label="background",stacked=True)
plt.legend(loc='upper right')
plt.show()
```

#### Part a

Compare ROC curves computed on the test versus training samples, in a single plot. Do you see a bias?

```python
from sklearn.metrics import roc_curve, roc_auc_score, auc
import matplotlib.pyplot as plt

f=Fisher.fit(X_Train,y_Train)
y_score1 = f.predict_proba(X_Test)[:,1]
y_score2 = f.predict_proba(X_Train)[:,1]

    
false_positive_rate1, true_positive_rate1, threshold1 = roc_curve(y_Test, y_score1)
false_positive_rate2, true_positive_rate2, threshold2 = roc_curve(y_Train, y_score2)

print('roc_auc_score1', roc_auc_score(y_Test, y_score1))

print('roc_auc_score2', roc_auc_score(y_Train, y_score2))
```

```python
auc_score_r1=auc(false_positive_rate1, true_positive_rate1)
auc_score_r2=auc(false_positive_rate2, true_positive_rate2)
```

```python
plt.figure()
plt.title('Receiver Operating Characteristic')
plt.plot(false_positive_rate1, true_positive_rate1, label='(area = {:.3f})'.format(auc_score_r1))
plt.plot(false_positive_rate2, true_positive_rate2, label='(area = {:.3f})'.format(auc_score_r2))
plt.plot([0, 1], [0, 1], 'k--')
plt.legend(loc='best')
plt.ylabel('True Positive Rate')
plt.xlabel('False Positive Rate')
plt.show()
```

```python
def compute_rate(d,bins=100):
    hist,bins_=np.histogram(d,bins=bins,density=True)
    R = np.cumsum(hist[::-1])[::-1] * (bins_[1]-bins_[0])
    return R,bins_
```

```python
def compare_significance(Test_sig, Test_bkg,X_Test,y_Test,model,
                         obs_name,
                         scenarios,bins=100,log=False):
        
    tpr,bins_sig = compute_rate(Test_sig[obs_name],bins=bins)
    fpr,bins_sig = compute_rate(Test_bkg[obs_name],bins=bins_sig)
    
    pred= model.predict_proba(X_Test)[:,1]
    FPR, TPR, _ = roc_curve(y_Test, pred)
    
    max_sigs=dict()
    table=list()
    
    for name, (n_sig_expected, n_bkg_expected) in scenarios.items():

        n_sig_expected_prime = n_sig_expected * TPR
        n_bkg_expected_prime = n_bkg_expected * FPR

        sig = n_sig_expected_prime/ np.sqrt(n_sig_expected_prime + n_bkg_expected_prime )
        plt.step(bins_sig[:-1],sig,label=name+" "+str((n_sig_expected, n_bkg_expected)))
    
        max_i=np.argmax(sig)
        max_sigs[name]=(max_i,n_sig_expected_prime[max_i],n_bkg_expected_prime[max_i],sig[max_i],bins_sig[max_i])
        table.append((name,n_sig_expected, n_bkg_expected, 
                      TPR[max_i],FPR[max_i],
                      n_sig_expected_prime[max_i],n_bkg_expected_prime[max_i],sig[max_i],bins_sig[max_i],max_i)
        )
    if log:
        plt.yscale("log")
    plt.legend()
    plt.show()
    
    display(HTML(tabulate.tabulate(table, tablefmt='html',
                                   headers=["Name",'N sig','N bkg',"TPR","FPR","N sig'","N bkg'",'sig','x_c',"bin i"])))
    return max_sigs
```

```python
scenarios = {
    "1": (10, 100),
    "2": (100, 1000),
    "3": (1000, 10000),
    "4": (10000, 100000)
}

for obs in FeatureNames:
    print(obs)
    _ = compare_significance(Test_sig, Test_bkg, X_Test,y_Test,model,obs, scenarios)
```

TPR/FPR/ROC/AUC are one way of assessing the quality of a classifier. Read about [Precision and Recall](https://en.wikipedia.org/wiki/Precision_and_recall), [Accuracy](https://en.wikipedia.org/wiki/Accuracy_and_precision), and [F-score](https://en.wikipedia.org/wiki/F-score).

#### Part b
Look through [model evaluation](https://scikit-learn.org/stable/modules/model_evaluation.html#) documentation. Using scikit-learns tools, compute TPR, FPR, ROC, AUC, Precision, Recall, F1 score, and accuracy for the method you selected in 4c above and each scenario. Make a nice table, which also includes the maximal significance. 


```python
from sklearn.metrics import roc_curve, auc, precision_score, recall_score, f1_score, accuracy_score

def compute_metrics(Test_sig, Test_bkg,X_Test,y_Test, obs_name, model):
    # Prepare the input data for prediction
    X_sig = Test_sig.drop("signal", axis=1)
    X_bkg = Test_bkg.drop("signal", axis=1)
    
    # Predict the probabilities using your trained model
    prob_sig = model.predict_proba(X_sig)[:, 1]
    prob_bkg = model.predict_proba(X_bkg)[:, 1]
    
    # Compute TPR and FPR
    pred= model.predict_proba(X_Test)[:,1]
    fpr, tpr, _ = roc_curve(y_Test, pred)
    
    # Compute ROC AUC
    roc_auc = auc(fpr, tpr)
    
    # Compute Precision, Recall, and F1 score
    y_pred = model.predict(X_Test)
    precision = precision_score(y_Test, y_pred)
    recall = recall_score(y_Test, y_pred)
    f1 = f1_score(y_Test, y_pred)
    
    # Compute Accuracy
    accuracy = accuracy_score(y_Test, y_pred)
    
    # Compute maximal significance
    n_sig_expected = Test_sig.shape[0]
    n_bkg_expected = Test_bkg.shape[0]
    n_sig_expected_prime = n_sig_expected * tpr
    n_bkg_expected_prime = n_bkg_expected * fpr
    sig = n_sig_expected_prime / np.sqrt(n_sig_expected_prime + n_bkg_expected_prime)
    max_i = np.argmax(sig)
    max_significance = sig[max_i]
    
    return {
        'TPR': tpr[max_i],
        'FPR': fpr[max_i],
        'ROC AUC': roc_auc,
        'Precision': precision,
        'Recall': recall,
        'F1 Score': f1,
        'Accuracy': accuracy,
        'Max Significance': max_significance
    }

metrics_table = []
for obs in FeatureNames:
    metrics = compute_metrics(Test_sig, Test_bkg,X_Test,y_Test, obs, model)
    metrics_table.append({
        'Observation': obs,
        'TPR': metrics['TPR'],
        'FPR': metrics['FPR'],
        'ROC AUC': metrics['ROC AUC'],
        'Precision': metrics['Precision'],
        'Recall': metrics['Recall'],
        'F1 Score': metrics['F1 Score'],
        'Accuracy': metrics['Accuracy'],
        'Max Significance': metrics['Max Significance']
    })

table_headers = ['Observation', 'TPR', 'FPR', 'ROC AUC', 'Precision', 'Recall', 'F1 Score', 'Accuracy', 'Max Significance']
table_rows = [[metrics['Observation'], metrics['TPR'], metrics['FPR'], metrics['ROC AUC'], metrics['Precision'], metrics['Recall'], metrics['F1 Score'], metrics['Accuracy'], metrics['Max Significance']] for metrics in metrics_table]

print(tabulate.tabulate(table_rows, tablefmt='github', headers=table_headers))
```
